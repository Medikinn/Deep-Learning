## 2) 첫 번째 머신러닝 프로그램




도미와 빙어 데이터를 합친 다음(49개) k-NN (k 최근접이웃) 방법으로 생선 종류를 분류해보자.

데이터들이 리스트로 이루어져있으므로 `+` 연산을 이용해서 결합할 수 있다.

<u>분류를 위한 사이킷런의 knn 클래스는 `neighbors` 모듈 안에 있는 `KNeighborsClassifier()`이다. 이 클래스를 이용하려면 feature 데이터가 들어있는  넘파이배열(또는 리스트) `X`와 target이 들어있는 배열(또는 리스트) `y`가 필요하다.</u>




이때 feature의 수가 `X`의 컬럼 수가 된다(Scikit-learn의 knn 분류 클래스에 대한 간단한 설명은 62쪽 참조).

따라서 우리가 분석할 데이터에는 feature가 `length, weight` 둘이므로 `X`는 2개의 컬럼을 갖는다. 즉 $ 49 \times 2 $ 행렬 모양의 리스트(2차원 리스트)를 만들면 된다. 교재 1장에서는 넘파이배열대신 리스트만 이용하고 있는데 2장에서는 리스트를 넘파이 배열로 바꿔서 분석한다.



리스트로 된 `length, weight` 두 feature 데이터를 2차원 리스트로 만드는 몇 가지 방법이 있는데 교재에서는 **`zip()` 함수와 list comprehension** 방법을 이용하고 있다.

이제 생선의 길이와 무게 두 feature를 묶어서 `fish_data`라는 이름의 2차원 리스트로 만들자.

```python
l=bl+sl
w=bw+sw
```

```python
data=[[l,w] for l,w in zip(l,w)]
```

<u>분류는 이처럼 정답(생선 이름)을 알고 있는 데이터로 모델을 만들어서 이름을 모르고 길이와 무게만 아는 새로운 생선이 나타났을 때 그 생선의 이름을 알아보려는 것이다. 이와 같이 정답이 있는 머신러닝을 **지도학습(supervised learnig)**이라고 한다.</u>

* **클래스**  

어떤 작업을 하기 위한 함수(method, 메서드)와 변수들(속성, attributes)을 모아놓은 프로그램을 **클래스(class)**라고 부르는데, 클래스는 객체지향프로그래밍(object oriented programming)의 핵심이다. 클래스를 만든는 방법이나 간단한 예들은 파이썬 교과서를 참고하고, 우리는 머신러닝을 위한 여러 클래스를 이용하는 방법을 중심으로 공부한다.

클래스는 흔히 붕어빵을 만들기 위한 붕어빵틀, 또는 어떤 실체를 구현하기 위한 설계도에 비유된다. 빵틀에 여러 가지 다른 재료를 넣어서 여러 종류의 빵을 만들 수 있는데 그런 빵들을 __인스턴스__, 또는 **객체(object)**라고 부른다.



* knn 분류를 위한 클래스

`KNeighborsClassifier` 클래스는 사이킷런의 `neighbors` 모듈 안에 들어있으므로 아래와 같이 불러서 객체를 만들 수 있다. 객체 이름은 `kn`이라고 붙이자.

객체 `kn`을 만들고 나면 클래스에 있는 속성이나 메서드들은 `kn.fit(), kn.score(), kn.predict(), kn.n_neighbors` 등과 같이 이용할 수 있다.
